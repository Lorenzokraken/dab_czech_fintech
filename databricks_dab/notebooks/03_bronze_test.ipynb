## 3. Verifica finale

Esegui questa cella dopo aver completato le sezioni 1 e 2.
Controlla che tutte le tabelle esistano e abbiano righe.

```python
expected = {
    "account":      4500,
    "client":       5369,
    "card":          892,
    "disp":         5369,
    "district":       77,
    "loan":          682,
    "order":        6471,
    "transactions":    40000,  # solo i 3 CSV caricati finora
}

for table, min_rows in expected.items():
    full_name = f"{CATALOG}.{BRONZE_SCHEMA}.{table}"
    assert spark.catalog.tableExists(full_name), f"❌ {table} non esiste"
    count = spark.table(full_name).count()
    cols  = len(spark.table(full_name).columns)
    assert count > 0, f"❌ {table} è vuota"
    print(f"{'✅' if count >= min_rows else '⚠️'} {table:<15} {count:>7} righe  {cols} colonne  (atteso ≥{min_rows})")

print("\n🎯 Bronze layer verificato.")
```
