```python
%sql
USE CATALOG czech_fintech;
USE SCHEMA bronze;
```

```python
%sql
CREATE OR REPLACE TABLE account (
  account_id    STRING,
  district_id   STRING,
  frequency     STRING,
  date          STRING,
  _ingestion_ts TIMESTAMP,
  _source_file  STRING
) USING DELTA;
```

```python
%sql
CREATE OR REPLACE TABLE client (
  client_id     STRING,
  birth_number  STRING,
  district_id   STRING,
  _ingestion_ts TIMESTAMP,
  _source_file  STRING
) USING DELTA;
```

```python
%sql
CREATE OR REPLACE TABLE card (
  card_id       STRING,
  disp_id       STRING,
  type          STRING,
  issued        STRING,
  _ingestion_ts TIMESTAMP,
  _source_file  STRING
) USING DELTA;
```

```python
%sql
CREATE OR REPLACE TABLE disp (
  disp_id       STRING,
  client_id     STRING,
  account_id    STRING,
  type          STRING,
  _ingestion_ts TIMESTAMP,
  _source_file  STRING
) USING DELTA;
```

```python
%sql
CREATE OR REPLACE TABLE district (
  A1  STRING,  -- district_id
  A2  STRING,  -- nome
  A3  STRING,  -- regione
  A4  STRING,  -- popolazione
  A5  STRING,  -- comuni <499
  A6  STRING,  -- comuni 500-1999
  A7  STRING,  -- comuni 2000-9999
  A8  STRING,  -- comuni >10000
  A9  STRING,  -- n° città
  A10 STRING,  -- % urbano
  A11 STRING,  -- salario medio
  A12 STRING,  -- disoccupazione 1995
  A13 STRING,  -- disoccupazione 1996
  A14 STRING,  -- imprenditori/1000ab
  A15 STRING,  -- crimini 1995
  A16 STRING,  -- crimini 1996
  _ingestion_ts TIMESTAMP,
  _source_file  STRING
) USING DELTA;
```

```python
%sql
CREATE OR REPLACE TABLE loan (
  loan_id       STRING,
  account_id    STRING,
  date          STRING,
  amount        STRING,
  duration      STRING,
  payments      STRING,
  status        STRING,
  _ingestion_ts TIMESTAMP,
  _source_file  STRING
) USING DELTA;
```

```python
%sql
CREATE OR REPLACE TABLE `order` (
  order_id      STRING,
  account_id    STRING,
  bank_to       STRING,
  account_to    STRING,
  amount        STRING,
  k_symbol      STRING,
  _ingestion_ts TIMESTAMP,
  _source_file  STRING
) USING DELTA;
```

```python
%sql
CREATE OR REPLACE TABLE transactions (
  transaction_id  STRING,
  account_id      STRING,
  date            STRING,
  type            STRING,
  operation       STRING,
  amount          STRING,
  balance         STRING,
  k_symbol        STRING,
  bank            STRING,
  account         STRING,
  _ingestion_ts   TIMESTAMP,
  _source_file    STRING
) USING DELTA;
```
