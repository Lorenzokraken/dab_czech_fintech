```python
CATALOG = "czech_fintech"
SCHEMA = "bronze"
tables = ["transactions", "account", "client", "card", "disp", "district", "loan", "order"]
for t in tables:
    spark.sql(f"DROP TABLE IF EXISTS {CATALOG}.{SCHEMA}.{t}")
print("✅ Bronze tables dropped")
```
